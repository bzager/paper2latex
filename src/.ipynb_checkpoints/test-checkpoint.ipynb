```python
# Load image

imgs = loadAll(maxsize=None)
displayAll(imgs)

plt.show()
```

```python
# Segment image

img = imgs[0]

radius = int(0.001*np.sqrt(img.size)) # disk structuring element radius
#method = "open"
method = "erode"
sig = 3.0 # sigma for Gaussian smoothing

bimg,labels,props = segmentation(img,radius=radius,method=method,sig=sig)
symbols = [Symbol(region,img) for region in props[1:]]
displayAll([img,bimg])

plt.tight_layout(pad=0.01)
plt.show()
```
