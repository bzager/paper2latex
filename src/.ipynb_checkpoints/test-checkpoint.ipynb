```python
# Segment image

def testSegment(img,radius,method,sig,size,extra):
    bimg,labels,props = segmentation(img,radius=radius,method=method,sig=sig)
    symbols = [Symbol(region,img=None,size=size,extra=extra) for region in props[1:]]
    squares = [sym.square for sym in symbols]
    displayAll([img,bimg])
    displayAll(squares)
    print("Radius: "+str(radius))
    print(len(squares))
    
    return symbols
    
img = imgs[1]

radius = int(0.002*np.sqrt(img.size)) # disk structuring element radius
method = "open"
#method = "erode"
#method = "close"
#method = "dilate"
sig = 1.0 # sigma for Gaussian smoothing

size = 45 # size of resized images (size x size)
extra = 5 # extra padding around resized images (size+extra x size+extra)

%time symbols = testSegment(img,radius,method,sig,size,extra)

plt.subplots_adjust(wspace=0,hspace=0)
#plt.tight_layout(pad=0.01)
plt.show()
```

```python
# Feature extraction (PHOG)

cells = [5,10,20]

for sym in symbols:
    sym.calcPhog(cells=cells)
```

```python
for sym in symbols[:10]:
    sym.displayPhog()
        
plt.tight_layout()
plt.show()
```
