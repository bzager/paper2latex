```python
import sys
sys.dont_write_bytecode = True
import importlib

import tools
import segment
import symbol
importlib.reload(tools)
importlib.reload(segment)
importlib.reload(symbol)

from tools import *
from segment import *
from symbol import Symbol

import numpy as np
import matplotlib.pyplot as plt

#import matplotlib.gridspec as gridspec

plt.rcParams['image.cmap'] = 'gray' # good colormaps: prism,flag
plt.rcParams['figure.dpi'] = 150

%matplotlib inline
```

```python
# Load images

imgs = loadAll(maxsize=None)
displayAll(imgs)

plt.show()
```

```python
# Segment image

def testSegment(img,radius,method,sig,size,extra):
    bimg,labels,props = segmentation(img,radius=radius,method=method,sig=sig)
    symbols = [Symbol(region,img,size=size,extra=extra) for region in props[1:]]
    squares = [sym.square for sym in symbols]
    displayAll([img,bimg])
    displayAll(squares)
    print("Radius: "+str(radius))
    print(len(squares))
    
    return symbols

img = imgs[1]

radius = int(0.002*np.sqrt(img.size)) # disk structuring element radius
method = "open"
#method = "erode"
#method = "close"
#method = "dilate"
sig = 1.0 # sigma for Gaussian smoothing

size = 56 # size of resized images (size x size)
extra = 4 # extra padding around resized images (size+extra x size+extra)

%time symbols = testSegment(img,radius,method,sig,size,extra)

plt.subplots_adjust(wspace=0,hspace=0)
#plt.tight_layout(pad=0.01)
plt.show()
```

```python
# Feature extraction

ori = 8
cell = (5,5)
block= (1,1)

for sym in symbols:
    sym.calcHOG(orientations=ori,cell=cell,block=block)
    
hogs = [sym.hogImg for sym in symbols]    
displayAll(hogs)

plt.subplots_adjust(wspace=0,hspace=0)
plt.tight_layout(pad=0.01)
plt.show()
```
